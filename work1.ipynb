```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([5,7])
y=np.array([6,3])
plt.plot(x,y,lw=20)
plt.grid(axis='y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
xpoints=np.array([5,7,])
ypoints=np.array([3,4,])
plt.plot(xpoints,ypoints,'o')
plt.title("Title 1")
plt.xlabel("x label", color="blue")
plt.ylabel("y label" ,color="blue")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([1, 8, 1,8,1])
y2 = np.array([7, 1, 7, 1,7])

plt.plot(y1, marker = '*')
plt.plot(y2,linestyle = 'dashed')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([2, 8, 1, 10])
plt.title('GRAPH')
plt.plot(ypoints, marker = 'o', mec = 'black',mfc = 'yellow')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2,2,8,1,30,115,145,60,70,57,67])
y = np.array([100,105,84,105,40,30,10,70,40,200,60])
sizes = np.array([50,20,60,30,70,30,70,20,65,105,95])
color=np.array([20,60,80,50,90,70,30,100,110,130,65])
plt.scatter(x,y,s=sizes,c=color)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C"])
y = np.array([4, 8, 6])

plt.bar(x, y, color = "red",width=0.1)
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 20, 20])
mylabels = ["25", "35", "20", "20"]
myexplode = [0.2, 0, 0, 0]
mycolors =["black", "hotpink", "red", "yellow"]
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True,colors=mycolors)
plt.legend()
plt.show() 
```
